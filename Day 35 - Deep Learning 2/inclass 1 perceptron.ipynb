# $Perceptron$

A perceptron is the simplest form of a neural network used primarily for binary classification problems. 
- also called the Artificial Neuron
- It combines inputs with weights and applies an activation function to make decisions. 
- Perceptrons serve as the building blocks for more complex neural networks, such as multi-layer perceptrons (MLPs) and deep neural networks (DNNs). 
- The perceptron algorithm is often referred to as a single-layer perceptron, distinguishing it from multilayer perceptrons, which are more complex. 
- Overall, perceptrons are fundamental in supervised learning for binary classifiers.

Let's implement the, Frank Rosenblatt suggested this algorithm:
[Learn about the algorithm](https://www.w3schools.com/ai/ai_perceptrons.asp)

1. Set a threshold value
2. Multiply all inputs with its weights
3. Sum all the results
4. Activate the output

```python
# simplest implementation of the rosenblat algorithm

import numpy as np

inputs = np.array([  1,    0,    0,    0,    1])
weights = np.array([0.7,  0.6,  0.8,  0.2,  0.1])

#1. Set a threshold value
threshold = 1.5

#2. Multiply all inputs with its weights
product = inputs * weights

#3. Sum all the results
weighted_sum = sum(product)

#4. Activate the output
def activation(weighted_sum):
    if weighted_sum > 1.5:
        return 'Yes, We Go!!'
    else:
        return "nope, Not going"
    
activation(weighted_sum)
```

#### Perceptron Learning
- perceptron can learn from examples through a process called training
- During training, the perceptron adjusts its **weights** based on observed errors
- using a learning algorithm such as the perceptron learning rule or a **backpropagation algorithm**
- The perceptron compares its output with the desired output and adjusts its weights accordingly
- aiming to minimize the error between the predicted and desired outputs

Perceptrons are often used as the building blocks for more complex neural networks, such as multi-layer perceptrons (MLPs) or deep neural networks (DNNs)

### Bias
- allows the perceptron to classify correctly even when **all input features are zero**
- a constant value, added to the weighted sum to shift the decision boundary
- Bias controls when the perceptron activates, independent of any input

Remind yourself of Linear Regression:

- in perceptron, we have `y = Wx + b`
