### Applications for Deep Learning

1. **Computer Vision** - 
- The ability of a computer to see
- We generally use **CNNs**: 
    - Object Detection: If there is an object in an image, detect its presence
    - Object Recognition: Recognize and classify the object in an image
    - Image Recognition: image detection then recognition - if you are dealing with faces
    - Self Driving Cars: Object detection/ recognition is going on at all times
        - we like to quantify these algorithms, by how many FPS we are getting Frames per Seconds

2. **Natural Language Processing** -
- The ability of a computer to understand, and generate Natural Language
- We generally use **RNNs**, **Transformers** e.g. from ChatGPT - (Chat Generative Pre-trained Transformer.)
    - Speech Detection/Recognition: if there is speech in a sound signal, detect and recognize it
    - Speech to Text: **Transcription** of speech into text
    - Text to speech: **TTS** you have text that you can turn to speech
    - Text to text: **Translation** of chaning the languages of the text
    - voice assistants, chat bots, LLMs

Other Significant Mentions:
- Large Lanugage Models: 
    - Large Language Models (LLMs) are advanced AI systems built on deep neural networks designed to process, understand, and generate human-like text.
    - Language models trained on huge amounts of data
    - https://en.wikipedia.org/wiki/Large_language_model
    
- Multi-Modal AI Models: models that can deal with text, images, videos etc
    - Multimodal AI refers to artificial intelligence systems that can process and integrate information from multiple modalities or sources of data. These modalities can include text, images, videos, audio, and other forms of sensory data.
    - https://www.geeksforgeeks.org/artificial-intelligence/what-is-multimodal-ai/

- Federated Learning: (collaborative learning)
    - Federated learning is a sub-field of machine learning
    - focuses on training models across multiple **decentralized** devices or nodes, 
    such as smartphones, IoT devices, and edge devices, without the need to transfer the data to a central server.
    - https://en.wikipedia.org/wiki/Federated_learning

- Neuromorphic Computing: (hardware that mimics human)

#### Comparison of **Deep Learning** vs **Machine Learning**:

Features | Machine Learning | Deep Learning
---------|------------------|--------------
Level of Complexity | relies on lesser number of features | Neural Networks learn directly from data 
Data Requirements | small and medium datasets | Require Large datasets
Feature Engineering | Needed | not needed
Computation Power | regular CPUs, requires lesser power | GPUs, TPUs or NPUs to meet the computational demand
Algorithms | Decision trees, Regression models, Neighbours, | CNNs, RNNs, Transformers etc
Applications | Structured data, csv, columns etc | Unstructured data, images, speech etc
Training Time | lesser | high 
Interpretability | are generally interpretatble - decision tree, can even visualize the decision | Black Box - we do not exactly know how the model came up with a result
Performance | Good for small tasks, better than DL for limited datasets | Great for large complex datasets

- A Perceptron is the simplest form of a **neural network** that makes decisions by combining 
**inputs** with **weights** 
and applying an **activation function**.

- Mainly used for binary classification problems
