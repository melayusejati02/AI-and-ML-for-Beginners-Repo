##### Check PC Specs:
- We want to know weather our PC has GPU or not:
Check the specs of your computer:


```python
# pip install psutil gputil
```

```python
import platform
import psutil
import GPUtil

def get_machine_specs():
    print("=== System Information ===")
    print(f"OS: {platform.system()} {platform.release()}")
    print(f"Processor: {platform.processor()}")
    print(f"Physical Cores: {psutil.cpu_count(logical=False)}")
    print(f"Total Threads: {psutil.cpu_count(logical=True)}")

    # RAM Info
    virtual_mem = psutil.virtual_memory()
    print(f"Total RAM: {virtual_mem.total / (1024**3):.2f} GB")

    print("\n=== GPU Information ===")
    try:
        gpus = GPUtil.getGPUs()
        if not gpus:
            print("No dedicated NVIDIA GPU detected.")
        else:
            for i, gpu in enumerate(gpus):
                print(f"GPU {i}: {gpu.name}")
                print(f" - VRAM Total: {gpu.memoryTotal} MB")
                print(f" - GPU Temp: {gpu.temperature}°C")
                print(f" - Driver: {gpu.driver}")
    except Exception as e:
        print(f"Could not retrieve GPU info: {e}")

if __name__ == "__main__":
    get_machine_specs()
```
