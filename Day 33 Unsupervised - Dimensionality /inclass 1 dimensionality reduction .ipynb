# $Dimensionality Reduction$

- So far, i have told you that:
    - More the data we have --- your models will learn more
- 
- When we add more features to our datasets, our datasets become complicated as well
- 
- if we are targetting a column or a prediction e.g. predict species, we will assess which columns, or features
- species is corelated with steeply
- high corelation means, we will be able to predict species based on the highly corelated feature
- 
- Some features that will have some corelation with the target, and thus will confuse our models
- 
- We have observed that it is in nobody's intrest to have a huge number of features
- 
- Can lead to overfitting as well, becuase we are adding more noise for the model to learn from 
- 
- lot of columns/features ----------- means information will be more spread out in different columns --- sparse
- 
- more analysis is needed -------- more computation 
- 
- lots of features also means that --- it will be harder for the models to analyse columns ---- to predict a target


#### $Curse-of-Dimensionality$

- <increasing more data, leads to decreased performance>
- 
- think of it as adding more noise to confuse the model
- 
- 
- We perform Dimensionality Reduction to remove the redundant features, on only select the top highly useful, and corelated ones
