#### 3. Regularizers 
- Regularization is a technique used in machine learning to prevent **overfitting** by adding a penalty for complexity
- models performing good on training data, but bad on unseen data
- types:
    1. Lasso (L1 Reg)
    - adds a absolute value of magnitude ofcoefficient/     weight to the loss function
    - The more weight you are assigning to a input,the  higher the penalty
    - Lamda λ is the multiplier to the weight,  Coefficient Shrinkage
    2. Ridge (L2 Reg)
    - adds a square of magnitude of coefficient/weight  to the loss function
    - more importance, steeper the penalty
    - Lamda λ is the multiplier to the weight,  Coefficient Shrinkage
    - 
    - Adjusting the λ (Coefficient Shrinkage) -     Regularization Parameter that controls the strenght     of regularization

    3. Elastic Net Regression
    - is a combination of both L1 as well as L2     regularization


#### 4. Activation Functions:
- Activation functions **introduce non-linearity** in neural networks, enabling them to learn complex patterns and make accurate predictions
- 
- Most real-world data is non-linear, such as relationships between house prices and features like size or location.
- 
- Activation functions decide whether a neuron should be activated 
- They also make backpropagation possible by providing **gradients** for **weight updates**.

### 1. **Sigmoid Function** - logistic function
- Formula: 
$$f(x) = \frac{1}{1 + e^{-x}}$$

- Output range: 0 to 1
- Ideal for: Binary classification problems
- Pros: Smooth, differentiable, probabilistic interpretation
- Cons: Can suffer from vanishing gradients for large input values

````python
# sigmoid functcion
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return (1/(1+ np.exp(-z)))

# x = np.array([x for x in range(10)])
x = np.linspace(-10,10,100)

y = sigmoid(x)

plt.plot(x,y, color = 'b')
plt.grid()
````

### 2. **Tanh** (Hyperbolic Tangent) Function
- Formula: 

$$\tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}$$

- Output range: -1 to 1
- Use case: Hidden layers in networks
- Pros: Zero-centered, stronger gradients than sigmoid
- Cons: Still susceptible to vanishing gradients

````python
# Tanh functcion

def tanh(z):
    return ((np.exp(z) - np.exp(-z))/(np.exp(z) + np.exp(-z)))

# x = np.array([x for x in range(10)])
x = np.linspace(-10,10,100)

y = tanh(x)

plt.plot(x,y, color = 'r')
plt.grid()
````
