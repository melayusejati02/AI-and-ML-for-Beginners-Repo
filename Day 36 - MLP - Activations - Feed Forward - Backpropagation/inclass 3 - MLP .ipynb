```python
import numpy as np
from tensorflow.keras.datasets import (
                                        mnist, # digits
                                        # fashion_mnist, # 10 classes of clothing
                                        # cifar10, # 10 classes of objects, birds planes cars etc
                                        # cifar100, # 100 classes of objects
                                        # imdb, # text movie reviews, sentiment analysis
                                        # reuters, # For Text, multiclass classification
                                        # bosting_housing # dataset for regression 
                                       )
```

```python
(X_train, y_train) , (X_test, y_test) = mnist.load_data() # brings 70K images to us

# [(array of shape 60000, 28 x 28)]-----> 60K 2d arrays
# # [(array of shape 600000,)]
# [(array of shape 10000, 28 x 28)]-----> 60K 2d arrays
# # [(array of shape 100000,)]
```

```python
np.set_printoptions(linewidth=200)
print(X_train[10])
print(y_train[10])
```

- We have a labelled dataset
- each image with a label

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

img_num = 80
x = X_train[img_num] # 28 x 28
label = y_train[img_num]

sns.heatmap(x, cmap='YlGnBu')
# plt.imshow(x, cmap='YlGnBu')
plt.title(f"letter: {label}")
```

### Crop and zoom the image:
- [choose image] [ choose rows , choose columns] 
- after we choose the image, we are indexing the two dimensions ---> rows, columns
- choosing rows and columns can be done with slicing
- 

```python
x = X_train[10][3:9 , 8:18] # [choose image] [ choose rows , choose columns] 
sns.heatmap(x, cmap= 'gray', vmin=0, vmax=255) 
```

#### Observe many images at once?
- we can use subplots
- 

```python
img_num = 40
x = X_train[img_num] # 28 x 28
label = y_train[img_num]

# sns.heatmap(x, cmap='YlGnBu')
# plt.title(f"letter: {label}")
# plt.show()


plt.figure(figsize= (8,3))

# subplots
# the figure has 1 row, 2 columns, and this plot is the first plot
plt.subplot(1,3,1)
sns.heatmap(x, cmap='YlGnBu')
plt.title(f"letter: {label}")

plt.subplot(1,3,2)
sns.heatmap(x, cmap='YlGnBu')
plt.title(f"letter: {label}")

plt.subplot(1,3,3)
sns.heatmap(x, cmap='YlGnBu')
plt.title(f"letter: {label}")
plt.show()
```

```python
# observe the images

plt.figure(figsize= (12,16)) # width x height  # 4 x 4

for i in range(24):    
    # subplots
    plt.subplot(6,6, i + 1)
    sns.heatmap(X_train[i], cmap='gray', cbar= False, xticklabels= False, yticklabels=False)
    plt.title(f"letter: {y_train[i]}")
```
