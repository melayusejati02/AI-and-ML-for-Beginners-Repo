# $Natural-Language-Processing-(NLP)$
- subfield of AI, that focuses on interaction between computers and humans (natural language)
- Natural Language ---> language that humans speak
- it is not simple like computer languages
- 
- computer languages have well defined syntax
- sure natural languages have grammer
    - is that grammer followed all the times?
    - no. not all times
    - that is what makes natural language complex
- 
- humans say something, mean something totally else
- NLP ---> linguistics, computer science, AI

- 
- Challenges with NL:
    - NL always needs context:
        - Example:
        - `it is freexing here`
        - does it mean cold? temperature?
        - it is a request, asking someone to please close the window
    - 
    - Words with two meanings:
        - `bat`
        - `i saw a boy with a bat`
        - do you mean the boy had a bat (something to play with)
        - or do you mean he had a bat (bird)
    - 
        - eg. `I saw a man on a hill with a telescope`
        - did you see a man on a hill, using a telescope?
        - Or did you just see a man on a hill, and he was carrying a telescope?

    - Sarcasm
        - `Oh great a flat tire is exactly what i needed today`
    - Tone
        - `Boss: I would be grateful if you could provide an update that was due yesterday.`
        - `Boss: Could you please give me the update now? you were supposed to do that yesterday.`
    - Slang
        - informal ways of communication
        - words are not included in dictionary
    - Typing errors
        - hitting wrong keys
        - `i just wnted to wasah my car`
        - we as humans have no problem understanding, but for a machine these can be tough

### 2. Stop word Removal
- frequently occuring words that don't add too much meaning to the text
- eg. is, the, and, an
- articles: the, a, an
- prepositions: in, on, at
- conjunctions: and, or, but
- pronouns: I, you, he, she
- Auxillary verbs: is, am, are
- 

```python
text = text.lower().split()
len(text) # we are using 61 tokens
```

```python
stopwords = ['the', 'a', 'an',  'in', 'on', 'at', 'and', 'or', 'but', 'I', 'you', 'he', 'she', 'is', 'am', 'are']

# text ---> tokens
# tokens - stopwords ---> Clean text

clean_text = [word for word in text if word not in stopwords]
len(clean_text) # we only use 47
```

```python
# punctuation removal
# . , / : ; \n

newtext = text

newtext = newtext.replace(".", "")
newtext = newtext.replace(",", "")
newtext = newtext.replace(":", "")
newtext = newtext.replace(";", "")
newtext = newtext.replace("\n", "")

newtext = newtext.lower()

newtext = newtext.split()

newtext = [word for word in newtext if word not in stopwords]
newtext
```

### Exercise:

1. Ask the user for prompt
2. create a function that caculates the tokens in user input prompt
    - lowercase, tokenize
    - calcualte the initial number of tokens used
    - stop word removal
    - calculate new number of tokens
    - set_limit = 50
    - give user a warning if the limit is reached
        - implement it on the prompt
        - Say "here is the answer", or "say token limit reached, Recharge!!"
        
3. this function will take user's input as prompt

> 25 Minutes

- you can connect an LLM with your machine
- 
- you keep this validation to limit the number of tokens used 
    - if a user enters number of words more than set limit
        - then you do not prompt the LLM
        - and you ask the user to revise the prompt
        - give back to them their previous prompt
        - 
    - User revises the prompt
        - you apply the validation again
        - if set limit is not crossed
        - then you prompt LLM or you ask them to revise
        
        

```python
print("Please insert your prompt")
user_input = input("Ask me a question: ")
 
def prompt(user_input):
 
    # Original Prompt
    print("Initial prompt by user:")
    print("-"*30)
    print(f"{user_input}")
    print("="*30)
    
 
    # Lowercase
    newtext = user_input.lower()
 
    # Tokenize
    newtext_token = user_input.split()
 
    # Number of token
    newtext_number_of_token = len(newtext_token)
    
    print(f"Number of token {newtext_number_of_token}")
    print("="*30)
 
    # Stop word removal
    stopwords = ['the', 'a', 'an', 'in', 'on', 'at', 'and', 'or', 'but',
                 'i', 'you', 'he', 'she', 'is', 'am', 'are']
   
    clean_newtext = [word for word in newtext_token if word not in stopwords]
 
    # Calculate new tokens
    newtext_token_after_removal = len(clean_newtext)
    
    print(f"Number of token after removal of stop words: {newtext_token_after_removal}")
    print("="*30)
 
    # Set limit = 50
    set_limit = 50
 
    # warning
    if newtext_token_after_removal < set_limit:
        print("="*30)
        print("Looking for answers, wait for the response...")
        print("="*30)
        print(f"Here are the answer: ")
        print("="*30)
        print(f"{clean_newtext}")
        print("="*30)
 
    else:
        print("="*30)
        print("\nToken has reached limit. Suscribe our plan for further usage")
 
prompt(user_input)
```

### Stemming
- Reduce words to their root form by chopping off the suffixes
- running ---> run
- 
- we remove `affixes`
    - Prefixes
        - undo 
        - `un`
    - Suffixes
        - running 
        - `ing`

- Infixes
    - a word between two words
        - abso-`bloody`-lutely
- Circumfixes
    - words that cover words on both sides
        - `en`light`en`
        - 

- Examples:
    - running ---> run
    - happier ---> happi
    - quickly ---> quick
    - 
    - changing ---> chang
    - changed ---> chang
    - change ---> chang
    - 
    - studying ---> studi
    - studies ---> studi
    - study ---> studi 

    - 
    - Improve ---> Improv
    - Improving ---> Improv
    - Improvements ---> Improv
    - Improved ---> Improv
    - Improver ---> Improv
